<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">5. Model training, testing and tuning:

```python
#Importing libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#Training KNN classifier and finding accuracies for training and test data for different values of K

#Setup arrays to store training and test accuracies
neighbors = np.arange(1,20)
train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)
    
    #Fit the model
    knn.fit(X_train, y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)
    
    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(X_test, y_test) 
```

```python
#Generate plot
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
print(test_accuracy[8])
```

As seen from the plot above, we get maximum testing accuracy for K=9 (for a value of K which is not too small like 1 or too large)

```python
#Setup a knn classifier with k=9 neighbors

knn = KNeighborsClassifier(n_neighbors=9)
```

```python
#fit the model

knn.fit(X_train,y_train)
```

```python
#Get accuracy score
y_pred=knn.predict(X_test)

print(accuracy_score(y_test,y_pred))
```

<b>Classification Report

```python
#import classification_report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

As seen above, model shows a pretty good metrics for each of the classes:
1) Precision - A high value of precision indicates that model is correctly able to identify True positives, that is correctly identify classes as Normal, Type_H or Type_S. This shows accuracy of positive predictions.
2) Recall - A high value of Recall indicates that model is correctly able to identify high fraction of True positivites. In this case it is able to identify classes as Normal, Type_H and Type_S; and only a small percentage of instances were wrongly mapped.
3) F1-score - A high value shows that a large percentage of positive predictions were correct.

```python
#Automate the task of finding best values of K for KNN.

# creating odd list of K for KNN
myList = list(range(1,30))

# subsetting just the odd ones
neighbors = list(filter(lambda x: x % 2 != 0, myList))
```

```python
# empty list that will hold accuracy scores
ac_scores = []

# perform accuracy metrics for values from 1,3,5....19
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # predict the response
    y_pred = knn.predict(X_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print("The best value of K for KNN is %d" % optimal_k)
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#00b3e5;">4. Data pre-processing:

```python
#Checking target column balance
custdf['LoanOnCard'].value_counts()
```

As we can see the target column data is not balanced. It has 90% of data where LoanOnCard is 0. This can create a bias while training the model. Therefore, it needs to be rebalanced.

```python
#We will rebalance Target column by Up-Sampling minority class, i.e. 0 in this case

from sklearn.utils import resample #import resample module
```

```python
# Separate majority and minority classes
df_majority = custdf[custdf.LoanOnCard==0]
df_minority = custdf[custdf.LoanOnCard==1]
 
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=4500,    # to match majority class
                                 random_state=123) # reproducible results
 
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
 
# Display new class counts
df_upsampled.LoanOnCard.value_counts()
```

```python
#Segregating predictors vs target attributes

X=df_upsampled.drop(labels='LoanOnCard',axis=1)
y=df_upsampled['LoanOnCard']

X.head()
```

```python
y.value_counts()
```

```python
#Performing train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```
